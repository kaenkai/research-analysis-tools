# Machine learning (ML) notebook

```python
import pandas as pd
import numpy as np
import sklearn
from matplotlib import pyplot as plt

from src.lib import ThinFilmDatabase
```

```python
df = ThinFilmDatabase.load_df()

print('Available columns:', ', '.join(df.columns.to_numpy()))
x1 = np.log10(df['TL10_15':'TL10_25']['stoichiometry_x'].to_numpy())
x2 = df['TL10_15':'TL10_25']['oxygen_pressure'].to_numpy()
y = df['TL10_15':'TL10_25']['crystallite_size'].to_numpy()

plt.figure()

plt.subplot(121)
plt.scatter(x1, y)
plt.xlabel('Deviation form stoichiometry')
plt.ylabel('Crystallite size [nm]')
plt.ylim(8, 20)

fit = np.polyfit(x1, y, 3)
r2_score = sklearn.metrics.r2_score(y, np.poly1d(fit)(x1))
print(r2_score)
x = np.linspace(min(x1), max(x1), 20)
plt.plot(x, np.poly1d(fit)(x), 'r')

plt.subplot(122)
plt.scatter(x2, y)
plt.xlabel('Oxygen pressure')
plt.ylim(8, 20)

fit = np.polyfit(x2, y, 1)
r2_score = sklearn.metrics.r2_score(y, np.poly1d(fit)(x2))
print(r2_score)
x = np.linspace(min(x2), max(x2), 20)
plt.plot(x, np.poly1d(fit)(x), 'r')

plt.show()
```
